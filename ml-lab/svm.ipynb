```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([[-2,4],[4,1],[1,6],[2,4],[6,2]])

Y=np.array([-1,-1,1,1,1])

plt.scatter(X[:,0],X[:,1])
plt.show()

X=np.array(
    [[-2,4,1],[4,1,1],[1,6,1],[2,4,1],[6,2,1]]
)

def sum_sgd(X,Y):
  W=np.zeros(len(X[0]))
  eta=1
  epochs=10000

  for epoch in range(1,epochs):
    for i,x in enumerate(X):
      if Y[i]*np.dot(X[i],W)<1:
        W=W+eta*((X[i]*Y[i])+(-2*(1/epoch)*W))
      else:
        W=W+eta*(-2*(1/epoch)*W)
  return W
W=sum_sgd(X,Y)
print(W)
for i in X:
  val=np.dot(i,W)
  print(np.sign(val))
plt.scatter(X[:,0],X[:,1],s=50,c=Y)

x2=[W[0],W[1],-W[1],W[0]]
x3=[W[0],W[1],W[1],-W[0]]
x2x3=np.array([x2,x3])

print("x2x3:",x2x3)
x1,Y,u,v=zip(*x2x3)

print("x1:",x1)
print("Y:",Y)
print("u:",u)
print("v:",v)

plt.quiver(x1,Y,u,v,scale=10,color='green')
plt.show()
```
