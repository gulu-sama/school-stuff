```python
import numpy as np
y=[1,1,0,1]
px1=[0.8,0.4,0.2,0.9]
```

```python
def likelihood(X,Y):
    lh=1
    for i in range (len(Y)):
        lh*=(X[i]**Y[i])*((1-X[i])**(1-Y[i]))
    print(lh)
```

```python
print(likelihood(px1,y))
```

```python
y1=[1,1,0,1]
px2=[0.8,0.4,0.2,0.9]
def loglikelihood(X,Y):
    lh=1
    for i in range(len(Y)):
        loglikelihood=np.sum(Y[i]*np.log(X[i])+(1-Y[i])*np.log(1-X[i]))
    print(lh)
```

```python
print(loglikelihood(px2,y1))
```
